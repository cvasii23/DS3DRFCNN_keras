```python
path2 = label_in
for i in range(len(os.listdir(path2)) // t):
    print(t* i,t*i+1,t*i+2,t*i+3)
    im0 = imread(os.path.join(path2,os.listdir(path2)[i]))
    im1 = imread(os.path.join(path2,os.listdir(path2)[i+1]))
    im2 = imread(os.path.join(path2,os.listdir(path2)[i+2]))
    im3 = imread(os.path.join(path2,os.listdir(path2)[i+3]))
    out1 = np.array([resize(im0,(h,w)), resize(im1,(h,w)),resize(im2,(h,w)),resize(im3,(h,w))])
    np.save(os.path.join(label_out,'label_' + str(t * i) + '.npy'), out1) 
```
